# ------------------------聚类分析 ------------------------

## 1.数据预处理

```python
from sklearn.preprocessing import scale
mobike=pd.DataFrame(scale(df_new))
```

## 2.用肘部原则选择cluster数目

```python
from sklearn.cluster import KMeans
elbow=[]
for i in range(1,10): # 创建遍历，找到最合适的k值
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=101)
    kmeans.fit(mobike)
    elbow.append(kmeans.inertia_)
```

```python
sns.lineplot(range(1,10),elbow,color='blue')
plt.rcParams.update({'figure.figsize':(16,10),'figure.dpi':100})
plt.title('Elbow Method')
```

###### 可见选择cluster=3时候模型效果最好

## 3.模型可视化

```python
from sklearn import cluster
model=cluster.KMeans(n_clusters=4,random_state=100)
model.fit(mobike)
```

```python
df_new['cluster']=model.labels_
df_new.head()
```

```python
sns.scatterplot(y='age',x='timeduration',hue='cluster',data=df_new)
```

```python
df_new.groupby(['cluster'])['timeduration'].describe()
```

###### 0组平均时间15mins以内
###### 1组平均时间15-30mins以内

```python
df_new.groupby(['cluster'])['age'].describe()
```

## 4.评估模型效果

```python
from sklearn import metrics
mobike_cluster=model.fit_predict(mobike)
score=metrics.silhouette_score(mobike,mobike_cluster)
print(score)
```

## 5.导出分群结果到csv文件

```python
centers=pd.DataFrame(model.cluster_centers_)
centers.to_csv('center_mobike.csv')
```
